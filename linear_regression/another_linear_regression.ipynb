```python
import polars as pl
import altair as alt
import numpy as np

# Generate sample data
np.random.seed(0)  # For reproducibility
n = 100
length = np.random.rand(n) * 100  # Random lengths between 0 and 100
# Linear relationship with noise
price = 2.5 * length + (np.random.rand(n) * 20)

# Create a Polars DataFrame
df = pl.DataFrame({
    "length": length,
    "price": price
})

# Scatter plot
scatter = alt.Chart(df).mark_circle(size=60).encode(
    x='length',
    y='price',
    tooltip=['length', 'price']
)


# Regression line
line = scatter.transform_regression('length', 'price').mark_line(color='red')

# Combine scatter plot and regression line
chart = scatter + line

display(df)
# Show the chart
chart.properties(
    width=600,
    height=400,
    title="Linear Regression of Price vs. Length"
).interactive()
```
