```python
import pandas as pd
```

```python
df = pd.read_csv(r'coffee.csv')

df
```

#### Iterate over DataFrame rows as (index, Series) pairs.


```python
for index, data in df.iterrows():
    print(index)
    print(data)
    print('-' * 24, end='\n\n')
```

#### Iterate over DataFrame rows as namedtuples.


```python
for data in df.itertuples(name='coffee', index=True):
    print(data)
    print('-' * 62)
```
