```python
import pandas as pd
```

```python
df = pd.read_csv('Flavors.csv')
df.head(1)
```

#### Rename columns


```python
df = df.rename(columns=lambda col: col.lower().replace(' ', '_'))
df
```

#### Drop duplicates


```python
# Cara simpel
df.drop_duplicates(subset=['base_flavor', 'liked'])

# ===

# Cara intuisi (ribet)
tmp = df[['base_flavor', 'liked']].duplicated()
tmp = tmp.drop(tmp.loc[tmp == False].index)

df.drop(index=tmp.index.to_list())
```
