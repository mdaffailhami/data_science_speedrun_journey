```python
import numpy as np
import pandas as pd
import plotly.express as px
```

Jill designs solar panels as a hobby.

On April 1st, Jill's "Mark I" design begins generating power: 1 kj/day (kj: kilo jule)

On May 1st, her "Mark II" design begins generating 4 kj/day

1. What day is it when Jill's Mark II design has generated as much total energy as the Mark I design?
2. How much total energy have both generated by that day?
3. What would the solutions to (1) and (2) be if Mark II design generated 1 kj/day?


```python
mark1_power = 1
mark2_power = 4  # If it's 1 (3rd question), then it'll be infinity

mark1 = np.array((mark1_power * 31, 31))
mark2 = np.array((mark2_power, 31))

while mark2[0] < mark1[0]:
    mark1[0] += mark1_power
    mark1[1] += 1

    mark2[0] += mark2_power
    mark2[1] += 1
```

```python
def generate_line(axis: str):
    a = [i for i in range(-50, 50+1, 10)]
    b = [0 for i in range(-50, 50+1, 10)]

    return a + b if axis == 'X' else b + a
```

```python
def generate_line_labels():
    a = ['Power (kj)' for i in range(-50, 50+1, 10)]
    b = ['Day' for i in range(-50, 50+1, 10)]

    return a + b
```

```python
df = pd.DataFrame({
    'Power (kj)': generate_line('X') + list(np.linspace(0, mark1[0], 10)) + list(np.linspace(0, mark2[0], 10)),
    'Day': generate_line('Y') + list(np.linspace(0, mark1[1], 10)) + list(np.linspace(31, mark2[1], 10)),
    'Label': generate_line_labels() + ['Mark I' for i in range(10)] + ['Mark II' for i in range(10)]
})

fig = px.line(
    df, 'Power (kj)', 'Day',
    color='Label',
    width=800, height=600,
    range_x=(-50, 50), range_y=(-50, 50),
    markers=True
)

fig
```

### Answers

1. 40 Days
2. 40 kj
3. No solutions
