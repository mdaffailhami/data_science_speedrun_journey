## Basic Algebra

### 1. Variables and Expressions
- **Variables:** Symbols used to represent unknown values (e.g., $x$, $y$).
- **Expressions:** Combinations of variables and numbers using operations (e.g., $2x + 3$).

### 2. Operations with Algebraic Expressions
- **Addition/Subtraction:** Combine like terms (e.g., $2x + 3x = 5x$).
- **Multiplication:** Distribute and multiply coefficients (e.g., $2(x + 3) = 2x + 6$).
- **Division:** Simplify expressions by dividing coefficients (e.g., $\frac{6x}{3} = 2x$).

### 3. Equations and Inequalities
- **Equations:** Statements that two expressions are equal (e.g., $2x + 3 = 7$).
- **Solving Equations:** Isolate the variable (e.g., $2x + 3 = 7 \rightarrow 2x = 4 \rightarrow x = 2$).
- **Inequalities:** Statements that compare expressions (e.g., $2x + 3 < 7$).
- **Solving Inequalities:** Similar to solving equations, but with special rules for multiplying/dividing by negative numbers.

## Intermediate Algebra

### 4. Polynomials
- **Definition:** Expressions with multiple terms (e.g., $3x^2 + 2x - 5$).
- **Operations:** Addition, subtraction, multiplication, and division of polynomials.

### 5. Factoring
- **Greatest Common Factor (GCF):** Factor out the largest common factor (e.g., $6x^2 + 9x = 3x(2x + 3)$).
- **Trinomials:** Factoring expressions like $x^2 + 5x + 6 = (x + 2)(x + 3)$.
- **Difference of Squares:** Factoring expressions like $x^2 - 9 = (x + 3)(x - 3)$.

### 6. Rational Expressions
- **Simplifying:** Reduce fractions to simplest form (e.g., $\frac{6x}{9} = \frac{2x}{3}$).
- **Operations:** Addition, subtraction, multiplication, and division of rational expressions.

## Advanced Algebra

### 7. Quadratic Equations
- **Standard Form:** $ax^2 + bx + c = 0$.
- **Solving Methods:**
  - Factoring: $x^2 + 5x + 6 = 0 \rightarrow (x + 2)(x + 3) = 0$.
  - Completing the Square.
  - Quadratic Formula: $x = \frac{-b \pm \sqrt{b^2 - 4ac}}{2a}$.

### 8. Functions and Graphs
- **Definition:** A function is a relation where each input has a unique output.
- **Types of Functions:** Linear, quadratic, exponential, etc.
- **Graphing:** Plotting functions on the coordinate plane.

### 9. Systems of Equations
- **Solving Methods:**
  - Graphing: Finding the intersection of two lines.
  - Substitution: Solving one equation for a variable and substituting into the other.
  - Elimination: Adding or subtracting equations to eliminate a variable.

### 10. Exponents and Logarithms
- **Exponents:** Rules for multiplying, dividing, and raising powers to powers.
- **Logarithms:** The inverse of exponents, with properties for simplifying expressions.

### 11. Complex Numbers
- **Definition:** Numbers in the form $a + bi$, where $i = \sqrt{-1}$.
- **Operations:** Addition, subtraction, multiplication, and division of complex numbers.
