# Laws of Exponents

The laws of exponents, also known as the rules of exponents or properties of exponents, are fundamental principles that describe how to manipulate and simplify expressions involving exponents. Here are the primary laws of exponents along with examples:

1. **Product of Powers Rule**:

   $a^m \cdot a^n = a^{m+n}$
   
   $2^3 \cdot 2^4 = 2^{3+4} = 2^7 = 128$

   $===$

   $2^3 \cdot 2^4 = 8 \cdot 16 = 128$
   
   When multiplying two expressions with the same base, add the exponents.

2. **Quotient of Powers Rule**:

   $\frac{a^m}{a^n} = a^{m-n} \quad \text{for } a \neq 0$
   
   $\frac{5^6}{5^2} = 5^{6-2} = 5^4 = 625$

   $===$

   $\frac{5^6}{5^2} = \frac{15625}{25} = 625$
   
   When dividing two expressions with the same base, subtract the exponent of the denominator from the exponent of the numerator.

3. **Power of a Power Rule**:

   $(a^m)^n = a^{m \cdot n}$
   
   $(3^2)^3 = 3^{2 \cdot 3} = 3^6 = 729$

   $===$

   $(3^2)^3 = 9^3 = 729$
   
   When raising an exponent to another exponent, multiply the exponents.

4. **Power of a Product Rule**:

   $(ab)^n = a^n \cdot b^n$
   
   $(2 \cdot 3)^4 = 2^4 \cdot 3^4 = 16 \cdot 81 = 1296$

   $===$

   $(2 \cdot 3)^4 = 6^4 = 1296$
   
   When raising a product to an exponent, raise each factor to the exponent.

5. **Power of a Quotient Rule**:

   $\left( \frac{a}{b} \right)^n = \frac{a^n}{b^n} \quad \text{for } b \neq 0$
   
   $\left( \frac{4}{2} \right)^3 = \frac{4^3}{2^3} = \frac{64}{8} = 8$

   $===$

   $\left( \frac{4}{2} \right)^3 = 2^3 = 8$
   
   When raising a quotient to an exponent, raise both the numerator and the denominator to the exponent.

6. **Zero Exponent Rule**:

   $a^0 = 1 \quad \text{for } a \neq 0$
   
   $7^0 = 1$
   
   Any non-zero base raised to the zero power is equal to one.

7. **Negative Exponent Rule**:

   $a^{-n} = \frac{1}{a^n} \quad \text{for } a \neq 0$
   
   $4^{-2} = \frac{1}{4^2} = \frac{1}{16} = 0.0625$
   
   A negative exponent indicates the reciprocal of the base raised to the opposite positive exponent.

8. **Fractional Exponent Rule**:

   $a^{\frac{m}{n}} = \sqrt[n]{a^m} = \left( \sqrt[n]{a} \right)^m$

   $27^{\frac{2}{3}} = \sqrt[3]{27^2} = (\sqrt[3]{27})^2 = 3^2 = 9$

   $===$
   
   $27^{\frac{2}{3}} = \sqrt[3]{27^2} = \sqrt[3]{729} = 9$
   
   A fractional exponent represents a root; the numerator is the power, and the denominator is the root.

These rules are essential for simplifying expressions and solving equations involving exponents.
