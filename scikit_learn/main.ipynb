```python
import polars as pl
import altair as alt
from sklearn.datasets import fetch_california_housing
from sklearn.neighbors import KNeighborsRegressor
```

```python
california_housing_data = fetch_california_housing()

X = pl.DataFrame(
    california_housing_data.data[:100],
    schema=california_housing_data.feature_names
)
y = pl.Series(
    california_housing_data.target_names[0],
    california_housing_data.target[:100]
)

display(X, y)
```

```python
mod = KNeighborsRegressor()
mod.fit(X, y)
```

```python
pred = mod.predict(X)

df = pl.DataFrame({'pred': pred, 'y': y})

scatter = alt.Chart(df).mark_circle(color='blue').encode(
    x='pred',
    y='y',
    tooltip=['pred', 'y']
).properties(width=500).interactive()

scatter + scatter.transform_regression('pred', 'y').mark_line(color='red')
```
