```python
from sklearn.datasets import fetch_california_housing
from sklearn.neighbors import KNeighborsRegressor
from sklearn.linear_model import LinearRegression
import numpy as np
import polars as pl
import altair as alt
```

```python
alt.data_transformers.enable("vegafusion")
```

```python
X, y = fetch_california_housing(return_X_y=True)

print(X)
print('\n' + '=' * 20, end='\n\n')
print(y)
```

```python
X = X[:100]
y = y[:100]
print(f'{len(X)} | {len(y)}')
```

```python
mod = LinearRegression()
```

```python
mod.fit(X, y)
```

```python
pred = mod.predict(X)
```

```python
df = pl.DataFrame({'prediction': pred, 'predicted': y})
scatter = alt.Chart(df).mark_circle(color='red').encode(
    x='prediction',
    y='predicted'
).properties(width=500).interactive()

line = scatter.transform_regression(
    'prediction', 'predicted').mark_line(color='red')

scatter + line
```
